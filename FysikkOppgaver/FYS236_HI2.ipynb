```python
import numpy as np
import matplotlib.pyplot as plt
```

### Simulating the magnetic field from a ring of current

the expression below has been found to give the magnetic field at a point P on the axis of a ring of current of radius R and current I. The ring is in the xy-plane and the point P is at a height z above the center of the ring.

```python
mu0 = 4 * np.pi * 10**-7 # permeability of free space in T m/A
r_0 = 0.01               # radius of the ring in meters
b = 0.200 + r_0 / 2      # radius of the ring in meters
J = 5.5e6               # current density in A/m^2
A = np.pi*r_0**2         # area of the ring in m^2
I = J * A                # current in the ring in Amperes
B = lambda z: mu0 * I * b**2 / (2 * (b**2 + z**2)**(3/2))
z = np.linspace(-0.6, 0.6, 1000) # z-axis vector in meters

plt.plot(z, B(z))
plt.xlabel('z (m)')
plt.ylabel('B (T)')
plt.grid()
plt.title('B field along the z axis through a current-carrying ring')
plt.savefig('ring.png', dpi=300)
plt.show()

```

### Getting the data from COMSOL

```python
# We want to read in the data generated by COMSOL and compare it.

data = np.loadtxt('hand_in2_data.txt', skiprows=9)

# we only want to plot the z-component of each vector

plt.plot(data[:,0], data[:,1])
plt.xlabel('z (mm)')
plt.ylabel('B (T)')
plt.grid()
plt.title('COMSOL data for Bz along the z axis through a current-carrying ring')
plt.savefig('comsol.png', dpi=300)
plt.show()
```

### Comparing the two results

```python
plt.plot(z*1000, B(z), label='Analytical', color='green')
plt.plot(data[:,0]-600, data[:,1], label='COMSOL', color='red')
plt.xlabel('z (mm)')
plt.ylabel('B (T)')
plt.grid()
plt.legend()
plt.title('Analytical vs COMSOL data for Bz')
plt.savefig('comparison.png', dpi=300)
plt.show()
```
